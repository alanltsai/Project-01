```python
plt.figure(figsize=(30,15))

plt.plot(East_Asia["DATE"],East_Asia["TMAX"],label="East Asia")
plt.xlabel("Year")
plt.ylabel("Maximum Temperature (F)")
plt.plot(Europe["DATE"],Europe["TMAX"],label="Europe")
plt.plot(North_Africa["DATE"],North_Africa["TMAX"],label="North Africa")
plt.plot(Latin_America["DATE"],Latin_America["TMAX"],label="Latin America")
plt.plot(North_America["DATE"],North_America["TMAX"],label="North America")
plt.plot(South_Asia["DATE"],South_Asia["TMAX"],label="South Asia")
plt.legend(loc='upper left',prop={'size':15})
plt.savefig("temperature_trend.png")
```
