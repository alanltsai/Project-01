```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(30,15))
plt.style.use('seaborn-darkgrid')
plt.grid(True)
plt.plot(East_Asia["DATE"],East_Asia["TMAX"],label="East Asia")
plt.xlabel("Year",fontsize=20)
plt.ylabel("Maximum Temperature (F)",fontsize=20)
plt.title('Regional Temperatures (1970-2018)', color='red', fontsize=20, fontweight=5)
plt.plot(Europe["DATE"],Europe["TMAX"],label="Europe")
plt.plot(North_Africa["DATE"],North_Africa["TMAX"],label="North Africa")
plt.plot(Latin_America["DATE"],Latin_America["TMAX"],label="Latin America")
plt.plot(North_America["DATE"],North_America["TMAX"],label="North America")
plt.plot(South_Asia["DATE"],South_Asia["TMAX"],label="South Asia")
#plt.legend(loc='upper left',prop={'size':15})

plt.text(Europe['DATE'].tail(1), Europe['TMAX'].tail(1), 'Europe', size='medium', horizontalalignment='left',fontsize=20, color="orange")
plt.text(North_Africa['DATE'].tail(1), North_Africa['TMAX'].tail(1), 'North Africa', size='medium', horizontalalignment='left',fontsize=20,color="green")
plt.text(North_America['DATE'].tail(1)-2.3, North_America['TMAX'].tail(1)-1, 'North America', size='medium', horizontalalignment='left',fontsize=20, color="purple")
plt.text(Latin_America['DATE'].tail(1), Latin_America['TMAX'].tail(1), 'Latin America', size='medium', horizontalalignment='left',fontsize=20, color="red")
plt.text(East_Asia['DATE'].tail(1), East_Asia['TMAX'].tail(1), 'East Asia', size='medium', horizontalalignment='left',fontsize=20,color="blue")
plt.text(South_Asia['DATE'].tail(1), South_Asia['TMAX'].tail(1), 'South Asia', size='medium', horizontalalignment='left',fontsize=20,color="brown")

plt.savefig("temperature_trend.png")
```

```python
plt.figure(figsize=(30,15))
plt.style.use('seaborn-darkgrid')
plt.grid(True)

plt.plot(East_Asia["DATE"],East_Asia["PRCP"],label="East Asia")
plt.xlabel("Year", fontsize=20)
plt.ylabel("Precipitation", fontsize=20)
plt.title('Regional Precipitation (1970-2018)', color='red', fontsize=20, fontweight=5)
plt.plot(Europe["DATE"],Europe["PRCP"],label="Europe")
plt.plot(North_Africa["DATE"],North_Africa["PRCP"],label="North Africa")
#plt.plot(Latin_America["DATE"],Latin_America["PRCP"],label="Latin America")
plt.plot(North_America["DATE"],North_America["PRCP"],label="North America")
plt.plot(South_Asia["DATE"],South_Asia["PRCP"],label="South Asia")
#plt.legend(loc='upper left',prop={'size':15})

plt.text(Europe['DATE'].tail(1), Europe['PRCP'].tail(1), 'Europe', size='medium', horizontalalignment='left',fontsize=20, color="orange")
plt.text(1998, 4.09, 'North Africa', size='medium', horizontalalignment='left',fontsize=20,color="green")
plt.text(North_America['DATE'].tail(1)-2.3, North_America['PRCP'].tail(1)-1, 'North America', size='medium', horizontalalignment='left',fontsize=20, color="red")
#plt.text(Latin_America['DATE'].tail(1), Latin_America['PRCP'].tail(1), 'Latin America', size='medium', horizontalalignment='left',fontsize=20, color="red")
plt.text(East_Asia['DATE'].tail(1), East_Asia['PRCP'].tail(1), 'East Asia', size='medium', horizontalalignment='left',fontsize=20,color="blue")
plt.text(2004, 49.23, 'South Asia', size='medium', horizontalalignment='left',fontsize=20,color="purple")

plt.savefig("precipitation_trend.png")
```
