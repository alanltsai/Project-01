```python
import pandas as pd
import gmaps
from config import gkey
gmaps.configure(api_key = gkey)
```

```python
air_file = "Resources/Air Quality Data.csv"
air_data = pd.read_csv(air_file)

city_file = "Resources/country-capitals.csv"
city_data = pd.read_csv(city_file)
```

```python
air_data.head()
```

```python
air_data2 = air_data.loc[:, ["country", "iso3", "region", "name", "exp_lower", "exp_mean", "exp_upper",
                            "year4", "pollutant", "measure"]]
air_data2.head()
```

```python
city_data = city_data.rename(columns={"CountryName" : "country"})
city_data.head()
```

```python
merged_data = air_data2.merge(city_data, on="country", how="left")
```

```python
merged_data = merged_data.dropna(how="any")
merged_data
```

```python
region_df = merged_data.loc[merged_data["region"] == "region", :]
region_df
```

```python
region_1990_df = region_df.loc[region_df["year4"] == 1990, :]
region_2016_df = region_df.loc[region_df["year4"] == 2016, :]
region_1990_df
```

```python
locations_1990 = region_1990_df[["CapitalLatitude", "CapitalLongitude"]]
locations_2016 = region_2016_df[["CapitalLatitude", "CapitalLongitude"]]
exp_1990 = region_1990_df["exp_mean"]
exp_2016 = region_2016_df["exp_mean"]
```

```python
fig_1990 = gmaps.figure()
fig_2016 = gmaps.figure()
```

```python
heat_layer_1990 = gmaps.heatmap_layer(locations_1990, weights = exp_1990,
                                     dissipating = False, max_intensity=20, point_radius=3)
heat_layer_2016 = gmaps.heatmap_layer(locations_2016, weights = exp_2016,
                                     dissipating = False, max_intensity=20, point_radius=3)
```

```python
fig_1990.add_layer(heat_layer_1990)
fig_2016.add_layer(heat_layer_2016)
```
